```python
import easyocr
import os
from PIL import Image
import cv2
from matplotlib import pyplot as plt
import numpy as np

from spellchecker import SpellChecker
```

```python
image_folder = 'C:/Users/oyku_/Desktop/image-pro/im/recipe_pics'
output_file = 'ocr_easy_results.txt'
```

```python
reader = easyocr.Reader(['en'])
```

```python
for filename in os.listdir(image_folder):
    if filename.lower().endswith(('.png', '.jpeg', '.jpg')):
        result = reader.readtext(os.path.join(image_folder, filename))
        print(f"Results for {filename}:")
        for detection in result:
            print(f"Text: {detection[1]} - Confidence: {detection[2]}")

```

```python
for filename in os.listdir(image_folder):
    if filename.lower().endswith(('.png', '.jpeg', '.jpg')):
        # Perform OCR on the image
        result = reader.readtext(os.path.join(image_folder, filename))

        # Print the image filename and detected text in the desired format
        print(f"Reading {filename}: ", end="")
        extracted_texts = [detection[1] for detection in result]  # Collect only the detected texts
        print(" | ".join(extracted_texts))
```

```python
with open(output_file, 'w', encoding='utf-8') as file:
    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.png', '.jpeg', '.jpg')):
            result = reader.readtext(os.path.join(image_folder, filename))
            
            extracted_texts = [detection[1] for detection in result]  
            line = f"Reading {filename}: " + " | ".join(extracted_texts) + "\n"  
            
            file.write(line)
            print(line.strip())
```

```python
def calculate_spelling_accuracy(text_file):

    spell = SpellChecker()

    with open(text_file, 'r', encoding='utf-8') as f:
        text_content = f.read()

    words = text_content.split()

    misspelled = spell.unknown(words)

    total_words = len(words)
    correct_words = total_words - len(misspelled)
    accuracy = correct_words / total_words if total_words > 0 else 0

    print(f"Total Words: {total_words}")
    print(f"Correct Words: {correct_words}")
    print(f"Misspelled Words: {len(misspelled)}")
    print(f"Spelling Accuracy: {accuracy:.2%}")

    return accuracy

text_file = 'C:/Users/oyku_/Desktop/image-pro/im/image_pro/ocr_easy_results.txt'
calculate_spelling_accuracy(text_file)
```
